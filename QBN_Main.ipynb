##### **The Density Matrix**

From a mathematical point of view, the quantum state can be advantageously described by a self-adjoint operator on a Hilbert space; this operator is called the **density matrix** of the quantum system.

In the pure state case, this operator is just the orthogonal projection on a subspace of the Hilbert space, while the density matrix of a mixed state is:

$$\rho = \sum_{j} \alpha_j | \psi_j \rangle \langle \psi |  $$

It should be kept in mind that the density matrix describes a preparation procedure for an ensemble of quantum systems whose statistical properties correspond to the given preparation procedure.


```
def my_density( superposition ):
  density = np.zeros((4,4), dtype=complex)
  for i in range(0, 4):
    for j in range(0, 4):

      real = superposition[:][0][i].real * superposition[:][0][j].real
      imaginary = superposition[:][0][i].imag + conj(superposition[:][0][j]).imag
    
      density[i][j] = complex(real, imaginary)

  return Qobj(density)
```

#### The Wigner Function of a Density Matrix

The Wigner function is a quasiprobability distribution of a wavefunction $\psi(x)$ is a real-values distribution in phase space defined as:

$$W(x,k)=\frac{1}{\pi} \int_{- \infty}^{\infty} dy \psi(x-y) \psi^* (x + y) e^{2iky},$$

where $k=p/\hbar$ is the wavenumber, closely related to the momentum, but in units of inverse length. $W$ often makes it easier to interpret wavefunctions than simply plotting $\psi(x)$, especially when $\psi(x)$ is complex-valued. 

Time-dependent wavefunctions are often plotted as a Wigner distribuion, which makes it easier to track a particle as it moves though phase space. In the classical limit, the time-dependent Wigner distribution becomes the classical phase-space density that satisfies the Liouville equation.

For a quick and easy evaluation of the Wigner distribution, we approximate the wavefnction as piecewise constant using 

$$\psi(x) \approx \psi(x_{[x /\Delta]} = v_{[x/\Delta]/\sqrt{\Delta}}$$ 

where we have used the calculation grid spacing $$\Delta = a / (n_{max} + 1)$$ and the nearest-integer rounding function $[z] = round(z)$. This approximation will be valid as long as $| k | << \pi / \Delta$
